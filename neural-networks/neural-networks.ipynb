# neural-networks

## 1. The neuron

```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
class Neuron:
    def __init__(self, weight1, weight2, bias):
        self.weights_1 = weight1
        self.weights_2 = weight2
        self.bias = bias
    
    def feedforward(cls, x1, x2):
        x1_w1 = x1 * cls.weights_1
        x1_w2 = x2 * cls.weights_2
        x = x1_w1 + x1_w2 + cls.bias
        y = sigmoid(x)
        return y
```

```python
neuron = Neuron(0, 1, 4)
neuron.feedforward(2, 3)
```
